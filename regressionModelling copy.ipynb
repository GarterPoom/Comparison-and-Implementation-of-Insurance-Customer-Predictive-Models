```python
#import library that need to use 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('insurance.csv')
```

```python
df.head()
```

```python
#use for identification data type
df.dtypes
```

```python
#Finding missing values for each feature
df.isnull().sum()
```

```python
num_data = df[['age','bmi','children']]
cat_data = df[['sex','smoker','region']]
class_target = df[['charges']]
```

```python
num_data.head()
```

```python
cat_data.head()
```

```python
cat_data = pd.get_dummies(cat_data)
cat_data
```

```python
num_data = pd.concat([num_data, cat_data], axis=1)
num_data
```

```python
class_target.head()
```

```python
# Create count plots for all nominal features
fig, ax = plt.subplots(2, 2, figsize=(15, 7))
for i, subplot in zip(num_data.columns, ax.flatten()):
    sns.boxplot(num_data[i], ax=subplot)
```

```python
#defining features (X) and (Y)
X = num_data
Y = class_target
```

```python
print(X)
```

```python
print(Y)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=0)
```

```python
print("x_train: ", x_train.shape)
print("x_test: ", x_test.shape)
print("y_train: ", y_train.shape)
print("y_test: ", y_test.shape)
```

```python
#Linear Regression Training
from sklearn.linear_model import LinearRegression

LR = LinearRegression()

LR.fit(x_train, y_train)

linearRPredictions = LR.predict(x_test)
```

```python
#The MSE (Mean Squared Error) evaluate:
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test, LR)
print('MSE is'+ mse)
```
